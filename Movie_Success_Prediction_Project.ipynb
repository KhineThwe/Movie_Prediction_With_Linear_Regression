```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('movie_data.csv')

data = data.replace([np.inf, -np.inf], np.nan)
data = data.dropna()
data = data.reset_index()
X = DataFrame(data,columns=['budget'])
y = DataFrame(data,columns=['gross'])
plt.scatter(x=X,y=y,alpha=1)

plt.title("Title: Budget and Gross")
plt.xlabel("Budget")
plt.ylabel("Gross")
plt.ylim(0,800000000)#million 0 8 lone#limits y axis range
plt.xlim(0,400000000)#million#limits x axis range

model = LinearRegression()#Linear regression model
model.fit(X,y)#x y two parameters

plt.plot(y,model.predict(y),color='red',linewidth=3)

print("coe",model.coef_)#m value
print("intercept",model.intercept_)#b value
print("r",model.score(X,y))

#Analysis
#y  = mx +b
#y = 1.04999162* 200000000+ 12005242.9035392
#33,005,075.3035392 #30 million gross= profit
# probability = 0.42 / 42% 
plt.show()
```
